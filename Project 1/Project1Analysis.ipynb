# In this file, I will be analyzing the 10 functions and their fitness outputs by finding out the average, standard deviation, range, median, and the overal time in milliseconds

```python
# Author: Zack Lee
# import statements
import pandas as pd
import numpy as np
import csv
# format scientific notation
pd.options.display.float_format = "{:.1f}".format
```

```python
# get the data into a dataframe
raw_data_df = pd.read_csv('results.csv') 
raw_data_df
```

```python
# split the dataframe into 10 different dataframes, one for each function
schwefelFunction_df = raw_data_df.iloc[0:900]
dejongFunction_df = raw_data_df.iloc[900:1800]
rosenbrockFunction_df = raw_data_df.iloc[1800:2700]
rastriginFunction_df = raw_data_df.iloc[2700:3600]
griewankFunction_df = raw_data_df.iloc[3600:4500]
sineEnvelopeSineWaveFunction_df = raw_data_df.iloc[4500:5400]
stretchedVSineWaveFunction_df = raw_data_df.iloc[5400:6300]
ackleyOneFunction_df = raw_data_df.iloc[6300:7200]
ackleyTwoFunction_df = raw_data_df.iloc[7200:8100]
eggHolderFunction_df = raw_data_df.iloc[8100:9000]
```

# Create functions to find the desired statistical data

```python
# create a function to find the average fitness of a function based on the resulting fitness values
def average_fitness(df):
    return df["fitness"].mean()
    
# create a function to get the standard deviation
def fitness_standard_deviation(df):
    return df["fitness"].std()

# create a function to find the range
def fitness_range(df):
    return df["fitness"].max() - df["fitness"].min()

# create a function to find the median
def fitness_median(df):
    return df["fitness"].median()

# create a function to calculate the total time for each function 
def total_function_time(df):
    return df["runtime_ms"].sum()
```

# Call each dataframe with each statistical analysis function and put it into one final dataframe

```python
function_dfs = {
    "Schwefel": schwefelFunction_df,
    "1st De Jong": dejongFunction_df,
    "Rosenbrock": rosenbrockFunction_df,
    "Rastrigin": rastriginFunction_df,
    "Griewank": griewankFunction_df,
    "Sine Envelope Sine Wave": sineEnvelopeSineWaveFunction_df,
    "Stretched V Sine Wave": stretchedVSineWaveFunction_df,
    "Ackley's One": ackleyOneFunction_df,
    "Ackley's Two": ackleyTwoFunction_df,
    "Egg Holder": eggHolderFunction_df
}

summary_rows = []

for name, df in function_dfs.items():
    summary_rows.append({
        "Function": name,
        "Average": average_fitness(df),
        "Median": fitness_median(df),
        "SD": fitness_standard_deviation(df),
        "Range": fitness_range(df),
        "Total Runtime (MS)": total_function_time(df)
    })

summary_df = pd.DataFrame(summary_rows)
summary_df.round(1)
```
